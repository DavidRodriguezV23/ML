```python
import numpy as np
import scipy as sc
import matplotlib.pyplot as plt
```

```python
func = lambda th: np.sin(1 / 2 * th[0] ** 2 - 1 / 4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1]) #Funci√≥n anonima
# th = [x, y]

res = 100

_X = np.linspace(-2, 2, res) # De -2 a 2 y res valores
_Y = np.linspace(-2, 2, res)

_Z = np.zeros((res, res))

for ix, x in enumerate(_X): #indice y valor
    for iy, y in enumerate(_Y):
        _Z[iy, ix] = func([x, y])
```

```python
plt.contourf(_X, _Y, _Z, 100) # resoluci√≥n es el ultimo parametro, con f para una malla solida
plt.colorbar()

Theta = np.random.rand(2) * 4 - 2 #Rango 0 a 1 

_T = np.copy(Theta)

h = 0.001
lr = 0.001 #learning rate

plt.plot(Theta[0], Theta[1], "o", c = 'white')

grad = np.zeros(2)

for _ in range(10000):

    for it, th in enumerate(Theta):
        
        _T = np.copy(Theta)
        
        _T[it] = _T[it] + h
        deriv = (func(_T) - func(Theta)) / h
    
        grad[it] = deriv

    Theta =  Theta - lr * grad

    if _ % 10 == 0:
            plt.plot(Theta[0], Theta[1], "o", c = 'red')
    
plt.plot(Theta[0], Theta[1], "o", c = 'green')
plt.show()
```

```python
function = lambda th: np.cos(th[0]) * np.cos(th[1])

quantity = 300

_X2 = np.linspace(-5, 5, quantity)
_Y2 = np.linspace(-5, 5, quantity)

_Z2 = np.zeros((quantity, quantity))

for ix, x in enumerate(_X2):
    for iy, y in enumerate(_Y2):
        _Z2[ix, iy] = function([x, y])
```

```python
plt.contourf(_X2, _Y2, _Z2, 100)
plt.colorbar()

Theta2 = np.random.rand(2) * 8 - 4

plt.plot(Theta2[0], Theta2[1], 'o', c = 'green')

plt.show()

_T2 = np.copy(Theta2)

grad2 = np.zeros(2)

h = 0.001
lr2 = 0.001

plt.contourf(_X2, _Y2, _Z2, 100)
plt.colorbar()
plt.plot(Theta2[0], Theta2[1], 'o', c = 'green')

for _ in range(10000):
    
    for it, th in enumerate(Theta2):
        
        _T2 = np.copy(Theta2)
        
        _T2[it] = _T2[it] + h
        
        d = (function(_T2) - function(Theta2)) / h   
        
        grad2[it] = d
    
    Theta2 =  Theta2 - lr * grad2
    
    if _ % 10 == 0:
        plt.plot(Theta2[0], Theta2[1], 'o', c = 'white')
    
plt.plot(Theta2[0], Theta2[1], 'o', c = 'red')
plt.show()
```
