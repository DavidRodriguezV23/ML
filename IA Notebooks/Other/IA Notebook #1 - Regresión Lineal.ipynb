```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_boston
from sklearn import linear_model
```

```python
#Load the library / dataset
boston = load_boston()

boston.keys() #Show functions

print(boston.DESCR)
```

Formula de minimizar el error cuadratico medio (MCO)
$\beta = (X^{T}X)^{-1}X^{T}Y$
Mean squared error

```python
X = np.array(boston.data[:,5])#RM
Y = np.array(boston.target)

plt.scatter(X,Y, alpha = 0.3) #alpha = transparency

#adding a col filled with 1 for the independent term
X=np.array([np.ones(506),X]).T

B = np.linalg.inv(X.T @ X) @ X.T @ Y

plt.plot([4, 9],[B[0]+B[1] * 4, B[0]+B[1] * 9], c= "red") #Giving two points of B
```

```python
X2 = np.array(boston.data[:,6])#AGE

plt.scatter(X2,Y)

X2 = np.array([np.ones(506),X2]).T

T = np.linalg.inv(X2.T @ X2) @ X2.T @ Y

plt.plot([0,100],[T[0], T[0]+T[1]*100], c="black")
```
